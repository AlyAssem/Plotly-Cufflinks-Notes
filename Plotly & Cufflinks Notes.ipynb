# Plots
scatter, 
bar, 
box, 
spread, 
ratio, 
heatmap,

histogram, 
bubble.

```python
# Scatter plot.
df.iplot(kind='scatter',x='A',y='B',mode='markers',size=10)
```

```python
# Bar plot.
# use an aggregate function to plot the dataframe.
# Gets the sum of all the data points in each column.
df.sum().iplot(kind='bar')
```

```python
# Box Plot
# u can simply compare only 2 boxes together by pressin on the labels to hide some of the columns.
df.iplot(kind='box')
```

```python
# Spread plot.
# Here it plots the normal line plot for both columns and then the spread difference between them in each point.
df[['A','B']].iplot(kind='spread')
```

```python
# Histogram
# if u hover over the pins the 1st number is the starting point of the bin 2nd number is the end point and
# 3rd is the number of points in that bin.
df['A'].iplot(kind='hist')
```
